```python
#prints "dog" five times
print("dog"*5)

#remember why this won't work
#print("dog"+5)

#remember why this does work?
#this is called concatenation
print("dog"+"cat")
```

```python
#above, we repeated an action, printing "dog", 5 times
#in coding, when we do repeated actions, we call it a loop
for i in [1,2,3,4,5]:
    print("dog", end='')
```

```python
#i runs five times because the size of the list is 5
#for each iteration, i is set equal to the next element in the list
for i in [1,2,3,4,5]:
    print(i)
```

```python
#i is a variable initialized inside the code block created by the for loop
#you can do things with the value in this variable
for i in [1,2,3,4,5]:
    print(i,"* 8 =", i*8)
```

```python
#range makes a set of values starting with the first value and ending before the second
#range is useful when we don't want to type the whole list
#imagine when you want to do a loop 20 times, a hundred, more!
for i in range(1,6):
    print(i,"* 8 =", i*8)
```

```python
#if you are starting at zero, then you only need to put in the number that you want to end before
for i in range(6):
    print(i,"* 8 =", i*8)
```

```python
#if you don't want to count in steps of one, you can specify your intervals in the third value
#also, technically, if I just wanted to print 20 and nothing higher,
#the upper bound can be anything between 21 and 25 inclusive
#because 20 is will still be the largest number that's a multiple of 5
for i in range(0,21,5):
    print(i)
```

```python
#this also works when you want to count down
#remember that when you do this, the bigger number (the one you start with)
#needs to be before the smaller number (the one you end with)
for i in range(10,0,-1):
    print(i)
```

```python
#when you print range, you don't get the values in it
#instead, it just prints range()
print(range(1,10))

#to print the values generated by range, put it in list()
print(list(range(1,10)))

#list() can also convert a string into a list of characters
print(list("Hello World"))
```

```python
#String is pretty much a list of characters,
#so when you put a string instead of a list in a for loop,
#i will be set equal to the first character and so on
for i in "U!S!A!":
    print(i)
```

```python
import time #time is a library, will get into this later

#lets make a countdown clock!
#count down from ten to one
for i in range(10,0,-1):
    print(i) #print number
    time.sleep(1) #wait one second
print("Blast Off!") #break out of loop and then print a countdown message
```

```python
#you can initialize a list ahead of time and put the variable name inside the for loop
thingsILike = ["tv","sugar","coding"]
for i in thingsILike:
    print("I like",i)
```

```python
#sometimes you don't know how many times you want to loop something
#instead you loop until something happens
print("We're gonna do some loops! Type 'y' for yes and 'n' for no")
anotherLoop = input()

#in this case, we will keep looping as long as the user types 'y'
#the loop will break when the user types anything else
while(anotherLoop == 'y'):
    print("We did a loop. Do you want to do another?")
    anotherLoop = input()
print("Ok, no more loops!")
```

```python
#when you use a continue, it stops where it's at in the loop instructions
#and starts a new iteration from the top
for i in range(1,6):
    print() #because when i == 3, there is no new line
    print("Hello!", end=" ")
    if(i==3):
        continue
    print("How are you?")
```

```python
#when you use break, stops running the loop altogether
for i in range(1,6):
    print() #because when i == 3, there is no new line
    print("Hello!", end=" ")
    if(i==3):
        break
    print("How are you?")
```

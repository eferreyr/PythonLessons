```python
#making all those ball objects was pretty labor intensive because you added the attributes one by one
#but you can initialize an object pre-loaded with all the attributes you want
#lets make a new class (or blueprint for some objects) that describes animals
class Animal:
    #here I will put a method that takes in certain attributes as arguments
    #and then initializes (creates) an instance of Animal (an Animal object)
    #the method MUST be called __init__(insert parameters here), with 2 underscores on either side, short for initialize
    #it still needs to have self in the parameters, too
    def __init__(self, name, classification, numLegs, cute):
        #here, I take the name that was passed as an argument
        #and assign whatever value it has to the name attribute inside of the Animal object current being made
        self.name = name
        #an animal can be classified under mammal, fish, bird, insect, amphibian, or reptile
        self.classification = classification
        #number of legs the animal has
        self.numLegs = numLegs
        #True or False? The animal is cute
        self.cute = cute
        
    #now remember how when you tried to print an object instead of just one attribute,
    #you would get a bunch of gibberish? Well, technically, it wasn't gibberish
    #the first bit between the <> is where the object was initialized
    #after the dot, is the name of the object type (same as what will show up if you put the object in type())
    #then it tells you the address of where your object is being stored in HEX
    #but this is not very useful to me. 
    #I want to see what's inside of my object, but how does it know what to print, in what order?
    #I can create instructions as to how to print my object by using a method called _str_()
    def __str__(self):
        #create a string variable called msg (short for message)
        #we will be adding onto this string as we go
        msg = ""
        
        #I will start by intruducing the animal by saying what it is called
        #if the name of the animal starts with a vowel, I will use "an" instead of "a" when introducing it
        #i do this by turning the string that holds the name of the animal into a list
        #now that I have a list where every element in the list is a letter in the string saved to name variable
        #I will take the first letter of the string my finding the letter in the first index 
        #and saving it to a variable called firstLetter
        nameList = list(self.name)
        firstLetter = nameList[0]
        
        #check if the first letter is a vowel
        if(firstLetter == "a" or firstLetter == "e" 
           or firstLetter == "i" or firstLetter == "o" or firstLetter == "u"):
            #here is a reason why the second way to format strings is useful
            #I can add plain old strings onto msg no problem
            #but I can't add ("This is an",self.name) or ("This is an %s"%self.name)
            #because they aren't strings, just instructions to create one
            #this way to do string formatting is able to be saved to variables
            #and therefore can be added onto msg
            msg += f"This is an {self.name}\n"
        #if it doesn't start with a vowel, then just use "a"
        else:
            msg += f"This is a {self.name}\n"
            
        #next, lets give its classification
        msg += f"It is a kind of {self.classification}\n"
        
        #now I will say how many legs it has
        msg+= f"It has {self.numLegs} legs\n"
        
        #finally if it is cute, I will say so
        if(self.cute): #passes if cute == True
            #since I'm not technically inserting anything that needs to be formatted,
            #I can cut the \n and just make it a plain old string
            msg += "And it is very cute!\n"
            
        #however, if you have nothing nice to say, don't say anything at all
        #so if the animal isn't cute, i won't mention it
        
        #now it's time to take the msg and return it
        #this string that you accumulated is what will be printed
        return msg
    
        #notice how my _str_() method let me do more than just print out my attributes
        #I can decide the order they are printed (I could've mentioned the number of legs before the classification)
        #I added strings to introduce my stats
        #in fact, I created if statements to decide whether the string I print uses "a" or "an"
        #also, I didn't even end up printing the contents of cute
        #sure, I used the value of cute in my if statement, but I didn't actually print "True"
        #instead, I had a string ready for when cute was True and straight up didn't do anything if it was false
        
    #####################################################################################
    #don't look or add the stuff below until after you have visited the Burger and Shapes file
    #####################################################################################
    
    #all animals eat, so we can put something as generic as that as a method in Animal
    def eat(self):
        print("Animal ate\n")
        
    #all animals sleep to
    def sleep(self):
        print("Animal slept\n")
        
    #if can't, however, add methods for walking, flying, or swimming, because not all animals can do all three
    #in fact, very few can, but a dog can walk, a hawk can fly, and goldfish can swim
    #i can still create these methods but they need to be in a more specific place
```

```python
#now, when I create an Animal object, I don't explicitly call the _init_() method
#instead, when I initialize frog as an Animal object, I put the attributes inside the parenthesis
#make sure that the attributes are inserted in the correct order
#and remember that the "self" in the parameter list of _init() is implied and isn't included in the Animal parenthesis
frog = Animal("frog","amphibian",4,True)
iguana = Animal("iguana","reptile",4,False)
tarantula = Animal("tarantula","insect",8,False)
chicken = Animal("chicken","bird",2,True)

print(frog)
print(iguana)
print(tarantula)
print(chicken)
```

```python
#a goldfish is an animal that can swim, but not walk 
#so I can create a goldfish class that has a swim function and not a walking one
#but a goldfish is also an animal in general, that eats and sleeps
#these are methods that are created in the animal class that I don't want to repeat here or let go to waste
#after all, a goldfish is just a more specific kind of animal

#using a concept called INHERITANCE, I can create the new, more specific class called Goldfish while
#still using the methods in Animal. 
#Since Animal is the more general class, it's called the SUPERCLASS
#since Goldfish is partly derived from Animal, it's called the SUBCLASS or DERIVED CLASS
#you make it this way by creating a Goldfish class like you would any other class
#but then, in parenthesis (), you add the name of the class you are inheriting from
class Fish(Animal):
    #now, i can borrow the initializing method __init__() from the animal class and add my own new attributes too
    #i will add numFins or the number of fins because that is something that's useful to know about fish specifically
    #i also already know that the fish will be classified as fish and have zero legs
    #so I don't need it as a parameter in my initialization method
    def __init__(self,name,cute,numFins):
        #for the attributes that are the same as those in animal, i just have Animal initialize them
        #for the attributes I already know for sure, I can just fill that in
        Animal.__init__(self,name,"fish",0,cute)
        #for the new attributes, I initialize them here
        self.numFins = numFins
        
    #now, we have access to the functions provided in animal, but we can also create our own
    #swimmming is specific to fish, so lets add a function for it
    def swim(self):
        print("Fish swam\n")
      
    #the same way that I called the function __init__() for Animal
    #to use for Fish, I can call Animal's eat() to be my Fish's eat()
    def eat(self):
        Animal.eat(self)
     
    #same goes for sleep()
    def sleep(self):
        Animal.sleep(self)
        
    def __str__(self):
        #we already know that if we just use Animal's __str__() method, it will print...
        #This is a [insert type of fish]
        #It is a kind of fish
        #It has 0 legs
        #(and if the fish is cute which most are, it will also print...)
        #And it is very cute!
        
        #but what if I want to add more to it? 
        #i will save the message so far to msg
        msg = Animal.__str__(self)
        
        #then i add the additional string that talks about the number of fins at the end
        msg += f"Also, it has {self.numFins} fins!\n"
        
        #return this new message
        return msg
```

```python
#now let's create a Fish object
goldfish = Fish("goldfish",True,7) #i googled this, goldfish have 7 fins!

#so we can use the swim function
goldfish.swim()

#and the eat and sleep functions in Animal
goldfish.eat()
goldfish.sleep()

print(goldfish)
```

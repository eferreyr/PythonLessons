```python
class Burger:
    #in the Ball class, we created objects with just methods and added attributes later
    #in the Animal class, we used __init__() so we could give certain attributes and
    #an Animal object would be created with those specified attributes
    #But we can also have a default object where all the attributes are predetermined
    #it's like almost no work! and don't worry, you can change them later, if you want
    def __init__(self):
        #my burger starts out not cooked at all, raw, or at a cooked level of zero
        self.cooked_level = 0
        #list of condiments on the burger, currently none, but can be added to later
        self.condiments = []
        # level of cookedness corresponds with indexes of cooked descriptions
        self.cooked_string = ["raw", "rare", "medium", "well done", "burnt", "charcoal"]

    def __str__(self):
        msg = "" #empty string for message that will be accumulated below

        #user can theoretically decide to keep cooking the burger even after it's turned to charcoal
        #but once it's turned to charcoal, it's not turning into anything else
        #from now on, the burger is charcoal for however much longer you cook it
        #but every time the user chooses to keep cooking, cooked_level increments once
        #it doesn't stay at 5 where it would correspond with the index of "charcoal"
        #and there aren't more elements in the list such that if I looked at index 6 or greater
        #there would be something there to print
        #therefore, for anything of cooked_level 5 or higher, we say that it is the same as
        #cooked_level 5, the element in index 5 of cooked_string, or "charcoal"
        if(self.cooked_level >= 5):
            msg += f"Burger is {self.cooked_string[5]} " #added a space at the end
        #if it is less than 5, then it's perfectly okay to use cooked_level to search
        #for the corresponding string in cooked_string
        else:
            msg += f"Burger is {self.cooked_string[self.cooked_level]} " #added a space at the end

        #if the list of condiments is empty, then skip this if statement and return the message as is
        #if the list of condiments does have stuff in it, then its length should be greater than zero
        #if that is the case, then print out the elements in the condiment list
        if(len(self.condiments) > 0):
            #add the first element of the list
            #we add the first element differently than the rest because all the elements afterwards
            #will be separated by an "and", but not the first
            msg += f"with {self.condiments[0]} " #added a space at the end

            #now it is also possible that the user only wants one condiment
            #or has only entered one condiment so far.
            #therefore, there are no other condiments to print
            #so I check if the length of condiments is greater than 1, meaning that it has more than one element
            #if that is the case, then I can go ahead and start printing the other condiments in the list
            if(len(self.condiments) > 1):
                #create a range from 1 to the length of the condiment list
                #so that i will equal 1, the index of the second element in the list,
                #up until one less than the length of the list, which is the index of the last element of the list
                #sometimes this is the same number, this is ok and it will only run once
                for i in range(1,len(self.condiments)):
                    msg += f"and {self.condiments[i]} " #added a space at the end

        return msg

    #when called, will increment cooked_level by one
    def cook_more(self):
        self.cooked_level += 1

    #when called, will add given string to end of condiments list
    def add_condiment(self,cond):
        self.condiments.append(cond)
```

```python
print("Welcome to my barbeque! Let me make you a burger!")
burger = Burger() #create a Burger object
print(burger) #at this point, it should be its default: raw with nothing on it

#ask if the user would like the burger cooked. If the input is "y", then do one iteration of while loop
#prompt says to type "n" for no, but we really are only checking whether or not input is equal to "y"
#if input is anything else, it will fail the conditional and not run the while loop
#we will ask this question every time we iterate and every time they say yes,
#we cook the burger some more
while(input("Would you like me to cook it some more for ya?\n Type 'y' for yes or type 'n' for no!") == "y"):
    burger.cook_more() #we want burger to cook more
    print(burger)

#once the user exits the first while loop, we enter a second one, but this time
#we ask whether they want anything on their burger (ie condiments)
#while loop will keep on running as long as the user answers yes to wanting to add another condiment
while(input("Would you like anything else on your burger?\n Type 'y' for yes or type 'n' for no!") == "y"):
    #prompt user to enter the condiment they would like to add to their burger
    #whatever they enter is added through add condiment
    burger.add_condiment(input("What condiment would you like?"))
    print(burger)

print("Here's your burger! Enjoy!")
#exit program
```
